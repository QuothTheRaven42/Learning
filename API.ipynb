```python
import requests
```

```python
"""HTTP methods"""

# read a resource
requests.get(url)

# create a response
requests.post(url, data={...})

# update a resource
requests.put(url, data={...})

# delete a resource
requests.delete(url)
```

```python
"""URL Parameters"""

query_params = {'artist':'Deep Purple'}
requests.get(url, params=query_params)
```

```python
"""Headers"""

headers = {'accept': 'application/json'}
response = requests.get(url, headers=headers)
print(response.headers.get('content-type'))
```

```python
"""Status codes"""

response = requests.get(url)
print(response.status_code)
```

```python
"""Authentication"""

# Basic authentication
headers = {'Authorization': 'Basic 38bvuvfn8u3b9...'}

# API key/token authentication
headers = {'Authorization': 'Bearer 38bvuvfn8u3b9...'}
```

```python
"""Structured Data"""

# Requesting JSON formatted data
requests.get(url, headers={'accept': 'application/json'})

# Sending JSON formatted data
playlists = [{'Name': 'Favorites'}, {'Name': "Road Trip"}]
requests.post(url, json=playlists)
```

```python
"""Error handling"""

from requests.exceptions import ConnectionError, HTTPError

try:
    response = requests.get(url)
    response.raise_for_status()

except ConnectionError as conn_err:
    print(f'Connection Error! {conn_err}.')

except HTTPError as {http_err}:
    print(f'HTTP error occurred! {http_err}.')
```
