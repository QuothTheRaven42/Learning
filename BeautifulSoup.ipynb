```python
"""Parsing and Navigating HTML"""
soup = BeautifulSoup(response.text, "html.parser")

"""Once parsed, several ways to navigate"""
# By tag name
# Using 'find' - returns one matching tag
# Using 'final_all' - returns a list of matching tags

"""Navigating with CSS Selectors"""
# 'select' - returns a list of elements matching the CSS selector
"""Selector Cheatsheet"""
# select by ID of 'foo': #foo
# select by class of 'bar': .bar
# select children: div > p
# select descendants: div p
```

```python
"""Accessing data in elements"""
soup[0].get_text()      # method to retrieve inner text if it exists
soup.name               # retrieves the name of the tag
soup.attrs              # dictionary of attributes
```

```python
"""Navigating via tags"""
data = soup.body                                # entire body as list
data.contents[1]                                # [1] due to \n first in list
data.contents[1].next_sibling                   # move to next sibling of selected tag
soup.find(class_="super-special").parent        # move up to parent tag
```

```python
""""Navigating via searching"""
soup.find("h3").find_next_sibling()             # ignores the \n to move to next sibling
soup.find("h3").find_previous_sibling()         # same, but previous sibling
soup.find("h3").find_parent()                   # move up to parent
soup.find("h3").find_child()                    # move down to next child
```
