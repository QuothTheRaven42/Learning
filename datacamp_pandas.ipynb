```python
import pandas as pd

wine_df = pd.read_csv("winequality-red.csv", sep=";")
```

```python
wine_df.head()
```

```python
print(wine_df["pH"].count())
```

```python
print(wine_df["pH"].mean())
print(wine_df["pH"].median())
```

```python
print(wine_df["pH"].min())
print(wine_df["pH"].max())
```

```python
print(wine_df["fixed acidity"].min())
```

```python
print(wine_df["fixed acidity"].max())
```

```python
print(wine_df["fixed acidity"].mean())
print(wine_df["fixed acidity"].median())
```

```python
cat_groups = (wine_df.groupby("quality")
              .agg(
    wine_count=("pH","count"),
    alcohol_mean = ("alcohol","mean"),
    alcohol_min = ("alcohol","min"),
    alcohol_max = ("alcohol","max")
)
              .reset_index())
```

```python
print(cat_groups)
```

```python
weed_df = pd.read_csv("kushy-cannabis-dataset/versions/1/products-kushy_api.2017-11-14.csv")
```

```python
grouping = (weed_df
            .groupby(["strain","category"])
            .agg(
                 strain_count = ("strain", "count"),
                 name_count = ("name", "count"))
            .reset_index())
print(grouping)
```
