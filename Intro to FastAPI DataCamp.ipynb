```python
"""Using pydantic's BaseModel"""
# an interface to define request and response body schemas
from pydantic import BaseModel

class Review(BaseModel): # inherit BaseModel's methods and attributes
    num_stars: int
    text: str
    public: bool = False

class MovieReview(BaseModel):
    movie: str
    review: Review # nest Review in Movie Review
```

```python
"""PUT and DELETE operations"""
# referencing existing objects
# No ORM, app must map object to ID
# Database ID - unique identifier
# _id convention for database IDs
## review_id: Table 'reviews', column 'id'
## Same conventions in frameworks with ORM
from pydantic import BaseModel

class DbReview(BaseModel):
    movie: str
    num_stars: int
    text: str
    review_id: int # references database ID column of Reviews
```

```python
"""Handling a PUT operation"""
# PUT endpoint to update an existing movie review:
# Endpoint: '/reviews'
# Input: DbReview
# Output: DbReview (updated)
@app.put("/reviews", response_model=DbReview)
def update_review(review: DbReview):
    db_review = crud.update_review(review) # updates movie review in database
    return db_review # updated review
```

```python
"""Handling a DELETE operation"""
# DELETE endpoint to remove an existing movie review
# Endpoint: '/reviews'
# Input: DbReview
@app.delete("/reviews")
def delete_review(review: DbReview):
    crud.delete_review(review) # deletes review from database
    return {} # nothing because the data is gone
```
