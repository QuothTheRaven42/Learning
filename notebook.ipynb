```python
"""
What is a query string?
A way to pass data to the server as part of a GET request
http://www.example.com/search?KEY1=QUERYSTRING
"""
```

```python
url = "https://icanhazdadjoke.com/search?term=cat"

response = requests.get(url,
                        headers={
                            "Accept": "application/json"
                        })

data = response.json()
print(data["results"][0]["joke"])
```

```python
url = "https://icanhazdadjoke.com/search"

term = input("What do you want a joke about?")

response = requests.get(url,
    headers={
            "Accept": "application/json"},
    params={
            "term": term
            })

data = response.json()
print(data["results"][0]["joke"])
```

```python
import random
import requests
from termcolor import colored

url = "https://icanhazdadjoke.com/search"

print(colored(r"""
  _____            _        _       _               _____  ___   ___   ___
 |  __ \          | |      | |     | |             | ____|/ _ \ / _ \ / _ \
 | |  | | __ _  __| |      | | ___ | | _____ _ __  | |__ | | | | | | | | | |
 | |  | |/ _` |/ _` |  _   | |/ _ \| |/ / _ \ '__| |___ \| | | | | | | | | |
 | |__| | (_| | (_| | | |__| | (_) |   <  __/ |     ___) | |_| | |_| | |_| |
 |_____/ \__,_|\__,_|  \____/ \___/|_|\_\___|_|    |____/ \___/ \___/ \___/

            """,
                color="blue"))

term = input("Let me tell you a joke! What do you want a joke about? ")

response = requests.get(url,
    headers={
            "Accept": "application/json"},
    params={
            "term": term
            })

data = response.json()
results = data["results"]

if data["total_jokes"]==0:
    print(colored(f"I have no jokes about '{term}'. Try another topic.", "red"))
elif data["total_jokes"]==1:
    print(f"I've got one joke about '{term}'. Here it is:")
    print(colored(results[0]["joke"], "green"))
else:
    print(f"I've got {data["total_jokes"]} jokes about '{term}'. Here's one:")
    joke = random.randint(0, data["total_jokes"]-1) # randint is inclusive on the total items
    print(colored(results[joke]["joke"], "green"))

```
