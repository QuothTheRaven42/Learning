```python
# filled star

from turtle import *

t1 = Turtle()
t1.color('orange')
t1.width(5)

t1.begin_fill()
for i in range(5):
    t1.forward(150)
    t1.left(144)
t1.end_fill()

done()
```

```python
# hollow pentagon, all sides different colors

from turtle import *

t1 = Turtle()
t1.width(5)

t1.color('orange')
t1.forward(150)

t1.color('purple')
t1.left(72)
t1.forward(150)

t1.color('green')
t1.left(72)
t1.forward(150)

t1.color('red')
t1.left(72)
t1.forward(150)

t1.color('cyan')
t1.left(72)
t1.forward(150)

done()
```

```python
# 4 colored squares in different locations

from turtle import *

t2 = Turtle()

def square_form():
    t2.down()
    t2.begin_fill()
    for i in range(4):
        t2.forward(100)
        t2.left(90)
    t2.end_fill()
    t2.up()

t2.color('red')
square_form()

t2.goto(100,100)
t2.color('blue')
square_form()

t2.goto(200,200)
t2.color('green')
square_form()

t2.goto(300,300)
t2.color('orange')
square_form()

done()
```

```python
# functions for three shapes

from turtle import *

t3 = Turtle()

def make_star():
    t3.down()
    t3.color('green')
    t3.begin_fill()
    for i in range(5):
        t3.forward(150)
        t3.left(144)
    t3.end_fill()
    t3.up()

def make_circle():
    t3.down()
    t3.color('blue')
    t3.begin_fill()
    t3.circle(20)
    t3.end_fill()
    t3.up()

def make_square():
    t3.down()
    t3.color('red')
    t3.begin_fill()
    for i in range(4):
        t3.forward(100)
        t3.left(90)
    t3.end_fill()
    t3.up()

make_square()

t3.goto(200,200)
make_circle()
t3.goto(50,50)
make_star()

done()
```

```python
# 10 circles at random locations

from turtle import *
from random import randint

t4 = Turtle()

for drawing in range(10):
    x = randint(-300, 300)
    y = randint(-300, 300)
    radius = randint(10,100)

    t4.up()
    t4.goto(x,y)
    t4.down()

    t4.circle(radius)

done()
```

```python
# 3 filled triangles in a column, all different colors

from turtle import *

t5 = Turtle()

def make_triangle():
    t5.begin_fill()
    t5.down()
    for i in range(3):
        t5.forward(100)
        t5.left(120)
    t5.up()
    t5.end_fill()

t5.color('red')
make_triangle()

t5.goto(0, 87)
t5.color('green')
make_triangle()

t5.goto(0, 174)
t5.color('blue')
make_triangle()

done()
```

```python
class App:
    def __init__(self, num_users, storage, name):
        self.name = name.strip().lower()
        self.num_users = num_users
        self.storage = storage

    def login(self):
        if self.name == 'owner' and self.num_users > 0:
            print(f'Login successful, {self.name}.')
            print(f'Your storage amount is {self.storage} GB.')
        else:
            raise ValueError('Access Denied.')

    def increase_capacity(self, amount):
        self.amount = amount
        self.storage += int(self.amount)
        print(f'New storage: {self.storage} GB.')

owner = App(36, 256, ' owNer ')
owner.login()
owner.increase_capacity(128)

# brian = App(-1, 128, 'Brian')
# brian.login() # ValueError
```
