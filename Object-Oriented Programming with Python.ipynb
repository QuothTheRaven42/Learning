```python
# functions for three shapes

from turtle import *

t3 = Turtle()

def make_star():
    t3.down()
    t3.color('green')
    t3.begin_fill()
    for i in range(5):
        t3.forward(150)
        t3.left(144)
    t3.end_fill()
    t3.up()

def make_circle():
    t3.down()
    t3.color('blue')
    t3.begin_fill()
    t3.circle(20)
    t3.end_fill()
    t3.up()

def make_square():
    t3.down()
    t3.color('red')
    t3.begin_fill()
    for i in range(4):
        t3.forward(100)
        t3.left(90)
    t3.end_fill()
    t3.up()

make_square()

t3.goto(200,200)
make_circle()
t3.goto(50,50)
make_star()

done()
```

```python
# online course management system

class OnlineCourse:

    def __init__(self, course, instructor):
        self.course = course.lower()
        self.instructor = instructor.lower()
        self.students = []

    def enroll_student(self, name):
        self.students.append(name.lower())
        print(f'{name.title()} successfully enrolled in {self.course}.')

    def get_course_details(self):
        print(f'The course is {self.course}, and the instructor is {self.instructor.title()}.')
        print(f'There are {len(self.students)} students taking the course.')
        print('Student names:', ", ".join(self.students).title())

    def complete_course(self, name):
        if name in self.students:
            self.students.remove(name)
            print(f'{name.title()} has successfully completed {self.course}.')
        else:
            print(f"{name.title()} is not enrolled in this course.")

    def calculate_avg(self, grades: list):
        avg = sum(grades) / len(grades)
        print(f'The average grade is {round(avg)}.')


math = OnlineCourse('math', 'patrick')

math.enroll_student('jeff beck')
math.enroll_student('john conner')
math.enroll_student('april o\'Niel')
print(math.students)
math.calculate_avg([95, 72, 101, 62, 79, 91, 84, 73, 100])
math.get_course_details()
math.complete_course('jeff beck')
print(*math.students)
math.complete_course('john beck')
```

```python
class Travel:
    def __init__(self, country, month, type):
        self.country = country.lower()
        self.month = int(month)
        self.type = type.lower()
        self.price = 0

    def trip_info(self):
        if self.month >= 10 or self.month <= 3:
            print(f"You are going to {self.country} in the Winter. This is a {self.type} trip.")
        elif self.month > 3 and self.month < 10:
            print(f"You are going to {self.country} in the Summer. This is a {self.type} trip.")
        else:
            print("Invalid Input")

    def calc_cost(self, cost):
        costs = []
        while cost:
            self.price += cost
            costs.append(cost)
            cost = int(input("Enter another cost: "))

        advice = self.advice(self.price)
        inspect = self.list_inspect(costs)
        return advice, inspect

    def advice(self, num):
        if num < 500:
            print(f"Low Budget!")
        elif num >= 500 and num < 1500:
            print("Take a flight to anywhere...")
        else:
            print("Luxury Trip")

    def list_inspect(self, costs):
        less_than_ten = 0
        for i in costs:
            if i >= 10:
                less_than_ten += 1

        if less_than_ten <= 10:
            self.price += 100
            print(f"Updated price: ${self.price}")


if __name__ == '__main__':

    location = input("Enter a country: ")
    month = input("Enter a month: ")
    trip_type = input("Leisure or business? ")

    test = Travel(location, month, trip_type)
    test.trip_info()

    flight_cost = int(input("Enter flight cost: "))
    test.calc_cost(flight_cost)
```

```python
class Guest:
    def __init__(self, firstname, lastname, rank, age):
        self.firstname = firstname
        self.lastname = lastname
        self.rank = int(rank)
        self.age = int(age)

    def guest_info(self, all_guests: list):
        for guest in all_guests:
            print(guest.firstname, guest.lastname, guest.age)

    def loyalty_points(self, all_guests: list):
        gold_members = [guest.lastname for guest in all_guests if guest.rank >= 10]
        if gold_members:
            print("Gold Members:")
            for member in gold_members:
                print("\tGuest:", member)

    def guest_avg(self, all_guests: list):
        ages = [guest.age for guest in all_guests]
        print(f"Average customer age: {round(sum(ages) / len(ages))}")


all_guests = []
num_guests = int(input("Enter number of guests: "))
for i in range(num_guests):
    data = input("First Name/Last Name/Rank/Age: ")
    firstname, lastname, rank, age = data.split("/")
    guest = Guest(firstname, lastname, rank, age)
    all_guests.append(guest)


guest = all_guests[0]
guest.loyalty_points(all_guests)
guest.guest_info(all_guests)
guest.guest_avg(all_guests)
```

```python
class Superhero:
    def __init__(self, name, power):
        self.name = name
        self.power = power

    def use_power(self):
        print(f"{self.name} is using the power of {self.power}!")

    def intro(self):
        print(f"This is {self.name} with the power of {self.power}.")

    def save_the_day(self):
        print(f"{self.name} has saved the day!")

    def power_level(self):
        length = len(self.power)
        level = length * 10
        return level


class FlyingSuperhero(Superhero):
    def __init__(self, name, power, speed):
        super().__init__(name, power)
        self.speed = speed

    def use_power(self):
        print(f"{self.name} is flying at a speed of {self.speed} mph.")

    def calc_distance(self, flight_time):
        distance = self.speed * flight_time
        print(f"{self.name} flew {distance} miles in {flight_time} minutes.")

superman = FlyingSuperhero("Superman", 'laser vision, flight, and strength', 200)
superman.use_power()
superman.calc_distance(12)
superman.save_the_day()
```

```python
"""
Library Management System - manage book collection
add book, display all books, search by author, search by year range
"""
from abc import ABC, abstractmethod

class Book(ABC):
    def __init__(self, title, author, year):
        self.title = title
        self.author = author
        self.year = year

    @property
    @abstractmethod
    def book_info(self):
        return f"{self.title} was written by {self.author} in {self.year}."


class FictionBook(Book):
    def __init__(self, title, author, year, genre):
        super().__init__(title, author, year)
        self.genre = genre

    @property
    def book_info(self):
        return f"{self.title} is a {self.genre} book written by {self.author} in {self.year}."


class NonFictionBook(Book):
    def __init__(self, title, author, year, topic):
        super().__init__(title, author, year)
        self.topic = topic

    @property
    def book_info(self):
        return f"{self.title} is a {self.topic} book written by {self.author} in {self.year}."


class Library:
    def __init__(self):
        self.books = []

    def add_book(self, book):
        self.books.append(book)

    def display_books(self):
        for book in self.books:
            print(book.book_info)

    def search_by_author(self, author):
        found_books = []
        self.author = author
        for book in self.books:
            if book.author == author:
                found_books.append(book)
            if found_books:
                print(f"Books by {author}:")
                for book2 in found_books:
                    print(f"{book2.year} {book2.title}")

    def search_year(self, start, end):
        found_books = []
        for book in self.books:
            if start <= book.year <= end:
                found_books.append(book)
        if found_books:
            print(f"Books between {start} and {end}:")
            for book in found_books:
                print(f"{book.year} {book.title}")
        else:
            print(f"No books found in this time period.")
```

```python
from abc import ABC, abstractmethod

class Country:
    def __init__(self, name, capital, population):
        self.name = name.title()
        self.capital = capital.title()
        self.population = population

    @abstractmethod
    def get_info(self):
        return {
            "Country": self.name,
            "Capital": self.capital,
            "Population": self.population,
        }


class Developed(Country):
    def __init__(self, name, capital, population, gdp):
        super().__init__(name, capital, population)
        self.gpd = gdp

    def get_info(self):
        info = super().get_info()
        info["GDP"] = self.gpd
        return info


class Developing(Country):
    def __init__(self, name, capital, population, hdi):
        super().__init__(name, capital, population)
        self.hdi = hdi

    def get_info(self):
        info = super().get_info()
        info["HDI"] = self.hdi
        return info

class World:
    def __init__(self):
        self.countries = []

    def add_country(self, country):
        self.countries.append(country)

    def country_info(self, name):
        for country in self.countries:
            if country.name == name:
                return country.get_info()
        return None

```

```python
world = World()

canada = Developed("Canada", "Montreal", 250000000, 300000000)
india = Developing("India", "New Delhi", 32000000, 0.645)
china = Developing('china', "Beijing", 1444216107, 0.758)

world.add_country(china)
world.add_country(canada)
world.add_country(india)
world.country_info('India')

country_info = world.country_info('Canada')
if country_info:
    print("Country Info: ")
    for key, value in country_info.items():
        print(f"{key}: {value}")
else:
    print('Country not found.')
```

```python
class Animal:
    def __init__(self, name):
        self.name = name

    def sleep(self):
        print(f'{self.name} is sleeping.')

    def eating(self):
        print(f'{self.name} is eating food.')

class CanFly:
    def flying(self):
        print(f'{self.name} is flying.')


class CanWalk:
    def walking(self):
        print(f'{self.name} is walking.')


class Bird(Animal, CanFly):
    def __init__(self, name, color):
        super().__init__(name)
        self.color = color

    def eating(self):
        print(f'{self.name} is eating seed.')

    def info(self):
        print(f'Name:' , self.name)
        print(f'Color:' , self.color)

birdie = Bird('Jimmie', 'blue')
birdie.info()
birdie.flying()
birdie.sleep()
birdie.eating()
```

```python
class Vehicle:
    def __init__(self, make, model, year):
        self.make = make
        self.model = model
        self.year = year


class Electric:
    def __init__(self, battery_cap, travel_time):
        self.battery_cap = battery_cap
        self.travel_time = travel_time

    def get_range(self):
        return self.battery_cap * self.travel_time


class Gas:
    def __init__(self, fuel_cap, mpg):
        self.fuel_cap = fuel_cap
        self.mpg = mpg

    def get_range(self):
        return self.fuel_cap * self.mpg


class Hybrid(Vehicle, Electric, Gas):
    def __init__(self, make, model, year, battery_cap, travel_time, fuel_cap, mpg):
        Vehicle.__init__(self, make, model, year)
        Electric.__init__(self, battery_cap, travel_time)
        Gas.__init__(self, fuel_cap, mpg)

    def get_range(self):
        return (self.battery_cap * self.travel_time) + (self.fuel_cap * self.mpg)


car = Hybrid("Toyota", "Highlander", 2020, 5, 40, 15, 42)

car.get_range()
```

```python
class Account:
    def __init__(self, acct_num, balance):
        self.acct_num = acct_num
        self._balance = balance
        self.transactions = []

    def __str__(self):
        return f'This is account #{self.acct_num}.\nThe balance is ${round(self._balance, 2)}'

    def __repr__(self):
        return f'Acct #{self.acct_num}\nBalance: ${round(self._balance, 2)}\nTransactions: {self.transactions}'

    def __len__(self):
        return len(self.transactions)

    @property
    def balance(self):
        return self._balance

    @balance.setter
    def balance(self, value):
        if value < 0:
            raise ValueError('Amount cannot be negative.')
        difference = round(value - self._balance, 2)
        if difference != 0:                          # ← only record actual changes
            self.transactions.append(difference)
        self._balance = value
        print(f'New Balance: ${round(value, 2)}')

    @balance.deleter
    def balance(self):
        raise ValueError(f'The account balance cannot be deleted. Please make a withdrawal.')

    def deposit(self, amount):
        self._balance += amount
        print(f'Depositing ${amount} into account #{self.acct_num}.')
        self.transactions.append(round(amount, 2))
        print(f'New Balance: ${round(self._balance, 2)}')

    def withdraw(self, amount):
        if self._balance >= amount:
            self._balance -= amount
            print(f'Withdrawing -${amount} from account #{self.acct_num}.')
            self.transactions.append(-round(amount, 2))
            print(f'New Balance: ${round(self._balance, 2)}')
        else:
            print(f'${amount} is larger than the account balance of ${round(self._balance, 2)}.')


class InterestBearing:
    def __init__(self, interest):
        self.interest = interest

    def add_interest(self):
        interest_amount = self._balance * self.interest
        print(f'Adding ${round(interest_amount, 2)} to the balance for interest.')
        self.transactions.append(round(interest_amount, 2))
        self._balance += interest_amount
        print(f'New Balance: ${round(self._balance, 2)}')


class TransactionHistory:
    def transaction_history(self):
        print("\nTransaction History:")
        for transaction in self.transactions:
            print(transaction)
            print("---------------------")
        print(f"Current Balance: ${round(self._balance,2)}")


class Savings(Account, InterestBearing, TransactionHistory):
    def __init__(self, acct_num, balance, interest):
        Account.__init__(self, acct_num, balance)
        InterestBearing.__init__(self, interest)


savings = Savings(5465471, 150, 0.0159)
```

```python
savings.deposit(520)
savings.withdraw(1000)
savings.add_interest()
savings.transaction_history()
```
