```python
# 4 colored squares in different locations

from turtle import *

t2 = Turtle()

def square_form():
    t2.down()
    t2.begin_fill()
    for i in range(4):
        t2.forward(100)
        t2.left(90)
    t2.end_fill()
    t2.up()

t2.color('red')
square_form()

t2.goto(100,100)
t2.color('blue')
square_form()

t2.goto(200,200)
t2.color('green')
square_form()

t2.goto(300,300)
t2.color('orange')
square_form()

done()
```

```python
# functions for three shapes

from turtle import *

t3 = Turtle()

def make_star():
    t3.down()
    t3.color('green')
    t3.begin_fill()
    for i in range(5):
        t3.forward(150)
        t3.left(144)
    t3.end_fill()
    t3.up()

def make_circle():
    t3.down()
    t3.color('blue')
    t3.begin_fill()
    t3.circle(20)
    t3.end_fill()
    t3.up()

def make_square():
    t3.down()
    t3.color('red')
    t3.begin_fill()
    for i in range(4):
        t3.forward(100)
        t3.left(90)
    t3.end_fill()
    t3.up()

make_square()

t3.goto(200,200)
make_circle()
t3.goto(50,50)
make_star()

done()
```

```python
class App:
    def __init__(self, num_users, storage, name):
        self.name = name.strip().lower()
        self.num_users = num_users
        self.storage = storage

    def login(self):
        if self.name == 'owner' and self.num_users > 0:
            print(f'Login successful, {self.name}.')
            print(f'Your storage amount is {self.storage} GB.')
        else:
            raise ValueError('Access Denied.')

    def increase_capacity(self, amount):
        self.amount = amount
        self.storage += int(self.amount)
        print(f'New storage: {self.storage} GB.')

owner = App(36, 256, ' owNer ')
owner.login()
owner.increase_capacity(128)

# brian = App(-1, 128, 'Brian')
# brian.login() # ValueError
```

```python
# online course management system

class OnlineCourse:

    def __init__(self, course, instructor):
        self.course = course.lower()
        self.instructor = instructor.lower()
        self.students = []

    def enroll_student(self, name):
        self.students.append(name.lower())
        print(f'{name.title()} successfully enrolled in {self.course}.')

    def get_course_details(self):
        print(f'The course is {self.course}, and the instructor is {self.instructor.title()}.')
        print(f'There are {len(self.students)} students taking the course.')
        print('Student names:', ", ".join(self.students).title())

    def complete_course(self, name):
        if name in self.students:
            self.students.remove(name)
            print(f'{name.title()} has successfully completed {self.course}.')
        else:
            print(f"{name.title()} is not enrolled in this course.")

    def calculate_avg(self, grades: list):
        avg = sum(grades) / len(grades)
        print(f'The average grade is {round(avg)}.')


math = OnlineCourse('math', 'patrick')

math.enroll_student('jeff beck')
math.enroll_student('john conner')
math.enroll_student('april o\'Niel')
print(math.students)
math.calculate_avg([95, 72, 101, 62, 79, 91, 84, 73, 100])
math.get_course_details()
math.complete_course('jeff beck')
print(*math.students)
math.complete_course('john beck')
```

```python
class Travel:
    def __init__(self, country, month, type):
        self.country = country.lower()
        self.month = int(month)
        self.type = type.lower()
        self.price = 0

    def trip_info(self):
        if self.month >= 10 or self.month <= 3:
            print(f"You are going to {self.country} in the Winter. This is a {self.type} trip.")
        elif self.month > 3 and self.month < 10:
            print(f"You are going to {self.country} in the Summer. This is a {self.type} trip.")
        else:
            print("Invalid Input")

    def calc_cost(self, cost):
        costs = []
        while cost:
            self.price += cost
            costs.append(cost)
            cost = int(input("Enter another cost: "))

        advice = self.advice(self.price)
        inspect = self.list_inspect(costs)
        return advice, inspect

    def advice(self, num):
        if num < 500:
            print(f"Low Budget!")
        elif num >= 500 and num < 1500:
            print("Take a flight to anywhere...")
        else:
            print("Luxury Trip")

    def list_inspect(self, costs):
        less_than_ten = 0
        for i in costs:
            if i >= 10:
                less_than_ten += 1

        if less_than_ten <= 10:
            self.price += 100
            print(f"Updated price: ${self.price}")


if __name__ == '__main__':

    location = input("Enter a country: ")
    month = input("Enter a month: ")
    trip_type = input("Leisure or business? ")

    test = Travel(location, month, trip_type)
    test.trip_info()

    flight_cost = int(input("Enter flight cost: "))
    test.calc_cost(flight_cost)
```

```python
class Guest:
    def __init__(self, firstname, lastname, rank, age):
        self.firstname = firstname
        self.lastname = lastname
        self.rank = int(rank)
        self.age = int(age)

    def guest_info(self, all_guests: list):
        for guest in all_guests:
            print(guest.firstname, guest.lastname, guest.age)

    def loyalty_points(self, all_guests: list):
        gold_members = [guest.lastname for guest in all_guests if guest.rank >= 10]
        if gold_members:
            print("Gold Members:")
            for member in gold_members:
                print("\tGuest:", member)

    def guest_avg(self, all_guests: list):
        ages = [guest.age for guest in all_guests]
        print(f"Average customer age: {round(sum(ages) / len(ages))}")


all_guests = []
num_guests = int(input("Enter number of guests: "))
for i in range(num_guests):
    data = input("First Name/Last Name/Rank/Age: ")
    firstname, lastname, rank, age = data.split("/")
    guest = Guest(firstname, lastname, rank, age)
    all_guests.append(guest)


guest = all_guests[0]
guest.loyalty_points(all_guests)
guest.guest_info(all_guests)
guest.guest_avg(all_guests)
```

```python
class Superhero:
    def __init__(self, name, power):
        self.name = name
        self.power = power

    def use_power(self):
        print(f"{self.name} is using the power of {self.power}!")

    def intro(self):
        print(f"This is {self.name} with the power of {self.power}.")

    def save_the_day(self):
        print(f"{self.name} has saved the day!")

    def power_level(self):
        length = len(self.power)
        level = length * 10
        return level


class FlyingSuperhero(Superhero):
    def __init__(self, name, power, speed):
        super().__init__(name, power)
        self.speed = speed

    def use_power(self):
        print(f"{self.name} is flying at a speed of {self.speed} mph.")

    def calc_distance(self, flight_time):
        distance = self.speed * flight_time
        print(f"{self.name} flew {distance} miles in {flight_time} minutes.")

superman = FlyingSuperhero("Superman", 'laser vision, flight, and strength', 200)
superman.use_power()
superman.calc_distance(12)
superman.save_the_day()
```

```python
"""
Library Management System - manage book collection
add book, display all books, search by author, search by year range
"""
from abc import ABC, abstractmethod

class Book(ABC):
    def __init__(self, title, author, year):
        self.title = title
        self.author = author
        self.year = year

    @property
    @abstractmethod
    def book_info(self):
        return f"{self.title} was written by {self.author} in {self.year}."


class FictionBook(Book):
    def __init__(self, title, author, year, genre):
        super().__init__(title, author, year)
        self.genre = genre

    @property
    def book_info(self):
        return f"{self.title} is a {self.genre} book written by {self.author} in {self.year}."


class NonFictionBook(Book):
    def __init__(self, title, author, year, topic):
        super().__init__(title, author, year)
        self.topic = topic

    @property
    def book_info(self):
        return f"{self.title} is a {self.topic} book written by {self.author} in {self.year}."


class Library:
    def __init__(self):
        self.books = []

    def add_book(self, book):
        self.books.append(book)

    def display_books(self):
        for book in self.books:
            print(book.book_info)

    def search_by_author(self, author):
        found_books = []
        self.author = author
        for book in self.books:
            if book.author == author:
                found_books.append(book)
            if found_books:
                print(f"Books by {author}:")
                for book2 in found_books:
                    print(f"{book2.year} {book2.title}")

    def search_year(self, start, end):
        found_books = []
        for book in self.books:
            if start <= book.year <= end:
                found_books.append(book)
        if found_books:
            print(f"Books between {start} and {end}:")
            for book in found_books:
                print(f"{book.year} {book.title}")
        else:
            print(f"No books found in this time period.")
```

```python
harry_potter = FictionBook("Harry Potter", "JK Rowling", 1999, 'fantasy')
harry_potter.book_info
```

```python
nat_geo = NonFictionBook("National Geographic", "John Berryman", 2022, "geography")
nat_geo.book_info
```

```python
library = Library()
library.add_book(harry_potter)
library.add_book(nat_geo)
library.display_books()
```

```python
library.search_by_author("John Berryman")
library.search_year(1999, 2022)
```
